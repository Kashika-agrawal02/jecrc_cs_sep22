# Types of graphs


from typing import Optional
from IPython.core.interactiveshell import SingletonConfigurable


1. plot-    line graph ,between two parameters
2. bar-     bar graph of single parameter(Optional cat_data)
3. hist-    histogram of a bunch of data i.e. frequency of single parameter
4. scatter- dot graph between two params
5. boxplot/box-statstical overview in form of box for a single param 
                        mean , outliners, qunatiles(25%,75%),IQR
                        (optional-categorical data)
6. stem-           more like bar graph ,lines instead of bars 
7. violinplot- violin shaped statistical overview graph 
                spread of data (more like hist)
                Single parameter (Optional cat-data)

```python
y=x**3
```

```python
plt.plot(x,y)
plt.title('plot graph')
plt.show() #for script 
```

```python
plt.bar(x,y)
plt.title('bar graph')
plt.show() #for script 
```

```python
plt.hist(y)
plt.title('hist graph')
plt.show() #for script 
```

```python
plt.scatter(x,y)
plt.title('scatter graph')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,20))
plt.show()   #for script 
```

```python
plt.boxplot(y)
plt.title('box graph')
plt.show() #for script 
```

```python
plt.stem(y,x)
plt.title('stemgraph')

plt.show()
```

```python
plt.violinplot(x)
plt.title('violin graph')
 
plt.show()
```

```python
y2=x**4
```

```python
plt.plot(x,y2)
plt.show()
plt.plot(x,y)
plt.show()
```

```python
#plt.plot(x, y, x, y2)
fig=plt.figure(figsize=(5,5))
plt.plot(x, y,lw=3,marker='o',markersize=15,markerfacecolor='gray')
plt.plot(x,y2,marker='v',markersize=10)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```
