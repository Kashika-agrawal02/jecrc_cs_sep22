### GRAPH Functions
1. plot                   - simple line plot
2. plot.area           - filled version of line plot
3. plot.bar  / plot.barh  - bar gra[h]
4. plot.box              --  box  plot
5. plot.density        --density line (smooth curved)
6. plot.hexbin        --hexagon version of scatter with density
7. plot.hist              --- histogram
8. plot.kde              ---- kernal density estimator (density graph)
10. plot.pie                  --pie chart of categorical data
11. plotscatter --            scatter graph

```python
fig=plt.figure(figsize=(20,20))                #remove 1st line and remove plt.subplot and add plt.show() at the end

#plt.subplot(4,3,1);
df2.plot();plt.title('Line Plot');plt.show();

df2.plot.area();plt.title('Area Plot');plt.show();
df2.plot.bar();plt.title('Bar Plot');plt.show();
df2.plot.box();plt.title('Box Plot');plt.show();  # boxplot
df2.plot.barh();plt.title('Bar  horizontal Plot');plt.show();
df2.plot.density();plt.title(' density Plot');plt.show();
df2.plot.kde();plt.title('KDE Plot');plt.show();

df2['a'].plot.hist();plt.title('Histogram Plot');plt.show();
df2.plot.scatter('a',y='b');plt.title('Scatter plot');plt.show();

df2.plot.hexbin(x='a' ,y='b' ,C='c' ,gridsize=6);plt.title('Hexbin plot');plt.show();
```

##### Categories
1. Categorical Plots (hue can be used | comparison with categorical data)
   a. count plot  (single param all)
   b. bar plot   (one categorical ,one numerical | calculate mean by default)  //
   c. box plot   (syntax same except estimator,single)
   d. violinplot    (single para )
   e. swarmplot
2. Distribution Plots
   a. distplot
   b. displot
   c. jointplot
   d. pairplot
   e. kdeplot
   f. rugplot
   g. scatterplot
3. Mtrix Plots 
   a. headtmap
   b. clustermap
   c. pairplot(same as above one)
4. Regression plots
   a. lmplot
   b. regplot
5. Grid Plots
   a. pairgrid
   b. Jointgrid
   e. facetgrid

```python
sns.countplot(x='smoker',data=tips)
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.mean,hue='time')  
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker');plt.show()
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker') ;plt.show()  
sns.swarmplot(x='day',y='total_bill',data=tips,hue='smoker');plt.show()
```
